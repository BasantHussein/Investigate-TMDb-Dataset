### Data Cleaning 

```python
#drop the columns that are not useful,and check by printing few lines
drop_col=['imdb_id','homepage','tagline','keywords']
df.drop(drop_col, axis = 1, inplace = True, errors = 'ignore')
df.dropna(inplace=True)
df.head()
```

```python
#drop duplicated, and then check
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
#check
df.info()
```

```python
#covert the release date datatype to datetime then check
df['release_date']=df['release_date'].astype('datetime64[ns]')
df.info()
```

## Exploratory Data Analysis

> In this section we will answer the questions in the introduction section, and visualize them using histogram,scatter plots. 

### what is the budget and the revenue of the associated movie in terms of 2010 dollars, accounting for inflation over time?

```python
#explore the budget accounting for inflation over time using histogram
df['budget_adj'].hist(color='blue',alpha=0.3)
```

#### After exploring the budget of accounting for inflation over time we found it skewed to the right,at the starting was maximum and then decreases.

```python
#explore the revenue accounting for inflation over time using histogram
df['revenue_adj'].hist(color='red',alpha=0.3)
```

#### After exploring the revenue of accounting for inflation over time we found it skewed to the right,at the starting was maximum and then decreases.

### what is the properties of the top 5 movies with the highest revenues?

```python
#make a dataframe with the sorted revenue column descending, then take the original data and put it in the new sorted dataframe,then print the top 5
df_sorted = pd.DataFrame(df['revenue'].sort_values(ascending=False))
for i in df.columns:
    df_sorted[i] = df[i]
df_sorted.head(5)
```

```python
#represent the relation between revenue and budget,runtime,popularity,vote_average using scatter plot
fig, axes = plt.subplots(2,2,figsize = (16,6))
df_sorted.plot(x='revenue', y='budget',color='c',kind='scatter',ax=axes[0][0])
df_sorted.plot(x='revenue', y='runtime',color='c',kind='scatter',ax=axes[0][1])
df_sorted.plot(x='revenue', y='popularity',color='c',kind='scatter',ax=axes[1][0])
df_sorted.plot(x='revenue', y='vote_average',color='c',kind='scatter',ax=axes[1][1])
```

#### From thoses representation we can find :
#### 1-there is a positive correlation between revenue and budget,popularity.
#### 2-there is a weak or may be no correlation between revenue and runtime,vote_average.
